#### Outlier Detection and Removal

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination=0.05, random_state=0)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
clean = iso.fit_predict(x,y)
```

```python
clean
```

```python
x_clean = x[clean==1]
y_clean = y[clean==1]
```

```python
x_clean.shape
```

```python
y_clean.shape
```

#### Data Balancing / Resampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0)
```

```python
x_res, y_res  = ros.fit_resample(x_clean, y_clean)
```

```python
x_res.shape
```

```python
sns.countplot(x = y_res)
```

```python
y_res.value_counts()
```

#### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf = SelectKBest(k=5, score_func=chi2)
```

```python
skf.fit(x_res,y_res)
```

```python
skf.get_support()
```

```python
x = x_res.iloc[:,skf.get_support()]    
x
```

#### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
pd.DataFrame(x_scaled).describe()
```

#### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
x_scaled, y_res, random_state=0, stratify=y_res)
```

```python
x_train.shape
```

```python
x_test.shape
```

#### Hyperparameter Tuning

```python
params = {
    
    'criterion': ["gini", "entropy", "log_loss"],
    'splitter' : ["best", "random"],
    'random_state': [0,1,2,3,4]
}
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
grid = GridSearchCV(DecisionTreeClassifier(), param_grid=params, cv=5)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_estimator_
```

#### Build the model

```python
DTC =  grid.best_estimator_
```

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
DTC_pred = DTC.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, DTC_pred)
```

```python
accuracy_score(y_test, DTC_pred)
```

```python
print(classification_report(y_test, DTC_pred))
```

```python
from tkinter import *
```

```python
import tkinter.messagebox
import numpy as np
import matplotlib.pyplot as plt
import joblib
```

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
top = Tk()
top.title('RICE CLASS CATEGORIZATION SYSTEM')
top.geometry("1500x1000")
top.resizable(0, 0)

#Deserializatiom
classifier = joblib.load('classifier.model')
scaler = joblib.load('scaler.model')

def check():
    global a, b, c, d, e
    new = np.array([[a.get(), b.get(), c.get(), d.get(), e.get()]])
    new = pd.DataFrame({
        'Area' : [a.get()],
        'Perimeter' : [b.get()],
        'Major_Axis_Length' : [c.get()],
        'Minor_Axis_Length' : [d.get()],
        'Convex_Area' : [e.get()]
    })
    y_pred = classifier.predict(scaler.transform(new))
    if y_pred[0]== 'Cammeo':
        tkinter.messagebox.showinfo('Result', "Cammeo")
    elif y_pred[0] == 'Osmancik':
        tkinter.messagebox.showinfo('Result', "Osmancik")
        
a = DoubleVar()
b = DoubleVar()
c = DoubleVar()
d = DoubleVar()
e = DoubleVar()

Label(top, text="RICE CLASS CATEGORIZATION SYSTEM", 
      font=('Ubuntu', 25)).place(x=120,y=30)
Label(top, text="Area:", font=('courier', 16)).place(x=20,y=100)
Entry(top, textvariable=a, font=('courier', 16)).place(x=280,y=100)
Label(top, text="Perimeter:", font=('courier', 16)).place(x=20,y=150)
Entry(top, textvariable=b, font=('courier', 16)).place(x=280,y=150)
Label(top, text="Major_Axis_Length:", font=('courier', 16)).place(x=20,y=200)
Entry(top, textvariable=c, font=('courier', 16)).place(x=280,y=200)
Label(top, text="Minor_Axis_Length",font=('courier', 16)).place(x=20,y=250)
Entry(top, textvariable=d, font=('courier', 16)).place(x=280,y=250)
Label(top, text="Convex_Area", font=('courier', 16)).place(x=20,y=300)
Entry(top, textvariable=e, font=('courier', 16)).place(x=280,y=300)

Button(top, text="--Check--", 
       font=('courier', 16, 'bold'), command=check).place(x=250,y=400)

top.mainloop()
```
