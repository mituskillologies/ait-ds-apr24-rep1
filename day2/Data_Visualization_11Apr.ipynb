```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([12,23,34,45,51,62,78,83,89,95])
y = np.array([9,14,26,31,46,53,61,67,73,85])
z = np.array([96,89,83,78,71,65,54,46,32,26])
```

### Line Plot

```python
plt.xlabel('first value')
plt.ylabel('Second value')
plt.title('Line Plot')
plt.plot(x,y, marker = 'o', mfc = 'red', color = 'cyan', ms = 7,
        ls = '--');
plt.grid()
# markers = >, <, ^, o, *
```

```python
plt.plot(x,z);
```

```python
plt.xlabel('first value')
plt.ylabel('Second value')
plt.title('Line Plot')
plt.plot(x,y, marker = 'o', mfc = 'red', color = 'blue', ms = 7,
        ls = '--', label = '2023');
plt.plot(x,z, marker = 'o', mfc = 'yellow', color = 'black', ms = 7,
        ls = '--', label = '2024');
plt.grid()
plt.legend();
```

### subplot

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.xlabel('first value')
plt.ylabel('Second value')
plt.title('Line Plot')
plt.plot(x,y, marker = 'o', mfc = 'red', color = 'blue', ms = 7,
        ls = '--');
plt.grid()

plt.subplot(2,2,2)
plt.xlabel('first value')
plt.ylabel('Second value')
plt.title('Line Plot')
plt.plot(x,z, marker = 'o', mfc = 'yellow', color = 'black', ms = 7,
        ls = '--');
plt.grid()
```

### Bar Graph

```python
students = [56,67,63,45,49]
courses = ['C', 'Python', 'Java', 'R', 'Kotlin']
```

```python
plt.xlabel('Course Name')
plt.ylabel('No. of Students')
plt.title('Student Enrolment')
colors = ['cyan', 'purple', 'black', 'orange', 'blue']
plt.bar(courses, students, color = colors, width = 0.5,
       edgecolor = 'red', linewidth = 3, alpha = 1);
plt.grid()
```

### Scatter plot

```python
plt.xlabel('first value')
plt.ylabel('second value')
plt.title('Scatter Plot')
plt.scatter(x, y, marker = '*');
plt.grid()
```

### Pie chart

```python
plt.title('Pie Chart')
plt.pie(students, labels = courses, autopct = '%.2f%%', shadow= True,
       radius = 0.8);
```

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.xlabel('first value')
plt.ylabel('Second value')
plt.title('Line Plot')
plt.plot(x,y, marker = 'o', mfc = 'red', color = 'blue', ms = 7,
        ls = '--', label = '2023');
plt.plot(x,z, marker = 'o', mfc = 'yellow', color = 'black', ms = 7,
        ls = '--', label = '2024');
plt.grid()
plt.legend();

plt.subplot(2,2,2)
plt.xlabel('Course Name')
plt.ylabel('No. of Students')
plt.title('Student Enrolment')
colors = ['cyan', 'purple', 'black', 'orange', 'blue']
plt.bar(courses, students, color = colors, width = 0.5,
       edgecolor = 'red', linewidth = 3, alpha = 1);
plt.grid()

plt.subplot(2,2,3)
plt.xlabel('first value')
plt.ylabel('second value')
plt.title('Scatter Plot')
plt.scatter(x, y, marker = '*');
plt.grid()

plt.subplot(2,2,4)
plt.title('Pie Chart')
plt.pie(students, labels = courses, autopct = '%.2f%%', shadow= True,
       radius = 0.8);

plt.savefig('Myplot.png')
```
